```python
import cv2
import os

#transformando foto em estilo "desenho"
def transformaremdesenho(arquivo, kernel_size):

    img = cv2.imread(f"imagens/{arquivo}") 

    imgpb= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #transforma a imagem em preto e branco

    imgpbinvert = cv2.bitwise_not(imgpb) #inverte as cores da imagem

    imgblur = cv2.GaussianBlur(imgpbinvert, (kernel_size, kernel_size), 0) #adiciona um "blur" na imagem (embaça). o kernel_size mede a "quantidade do filtro", com valores ímpares positivos

    imgblurinvert = cv2.bitwise_not(imgblur) #inverte a imagem novamente

    imgdesenho = cv2.divide(imgpb, imgblurinvert, scale=256.0) #coloca uma imagem em cima da outra e divide, como a máscara no photoshop


    cv2.imshow("Filhote", imgdesenho) #abre a imagem

    #esses dois comandos são para a imagem não entrar em loop 
    cv2.waitKey(0) 
    cv2.destroyAllWindows() 

    cv2.imwrite(f"imagem_editada/{arquivo}", imgdesenho) #salva a imagem 

lista_arquivos = os.listdir("imagens")

for arquivo in lista_arquivos:
    transformaremdesenho(arquivo, 55)

#esse código transforma todas as imagens da pasta "imagens" nesse estilo, o valor de "kernel_size" pode ser alterado (valores ímpares positivos)
```

```python
import numpy as np
import cv2
import time

#alterar o contraste da imagem

ESCAPE_KEY_ASCII = 27 #esc


def onChange(valor):    #mostra os valores alterados
    #print("Valor alterado", valor)
    pass

#imagem e cópia
img = cv2.imread("imagens/onedirection.jpg")
copiaimg = img.copy()


#janela
janelatitulo = "Ajuste de brilho e contraste"
cv2.namedWindow(janelatitulo)



#trackbar
cv2.createTrackbar("contraste", janelatitulo, 100, 100, onChange)

contrasteantes= 100
updatecontraste = False

conttempo = 0

while True:
    contrasteatual = cv2.getTrackbarPos("contraste", janelatitulo)
    
    #valor de contraste do trackbar alterado
    if contrasteantes != contrasteatual:
        updatecontraste = True
        conttempo = time.time()
        contrasteantes = contrasteatual
        
    #se for para atualizar o contraste e tiver passado 1 seg desde que o usuário mexeu no trackbar
    if updatecontraste == True and time.time() - conttempo > 1:
        #atualiza o contraste
        copiaimg = img.copy()


        height, width, channels = img.shape

        #para cada informação de cor, de cada pixel, atualiza o contraste
        for y in range (height):
            for x in range (width):
                for c in range (channels):
                    copiaimg[y][x][c] = copiaimg[y][x][c] *(contrasteatual/100)

        updatecontraste = False

    cv2.imshow(janelatitulo, copiaimg)


#quando a pessoa apertar "esc" a janela fecha
    k = cv2.waitKey (1) & 0xFF  
    if k == ESCAPE_KEY_ASCII:
        break

cv2.destroyAllWindows()

```
